```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
customer_churn_df = pd.read_csv("dataset/telecom_customer_churn.csv")
zipcode_population_df = pd.read_csv("dataset/telecom_zipcode_population.csv")
```

```python
# Exploratory Data Analysis (EDA)

# 1. Data Overview
print(customer_churn_df.head())
print(customer_churn_df.info())
print(customer_churn_df.describe())

# 2. Churn Rate Analysis
churn_rate = (customer_churn_df['Customer Status'] == 'Churned').mean()
print(f"Overall Churn Rate: {churn_rate*100:.2f}%")

# 3. Gender Distribution
sns.countplot(x='Gender', data=customer_churn_df, hue='Customer Status')
plt.title("Gender Distribution by Churn Status")
plt.show()

# 4. Age Distribution
sns.histplot(data=customer_churn_df, x='Age', hue='Customer Status', bins=20)
plt.title("Age Distribution by Churn Status")
plt.show()

# 5. Tenure Analysis
sns.boxplot(x='Customer Status', y='Tenure in Months', data=customer_churn_df)
plt.title("Tenure in Months by Churn Status")
plt.show()

# 6. Service Usage Analysis
sns.countplot(x='Internet Service', data=customer_churn_df, hue='Customer Status')
plt.title("Internet Service Usage by Churn Status")
plt.show()

# 7. Contract Type Analysis
sns.countplot(x='Contract', data=customer_churn_df, hue='Customer Status')
plt.title("Contract Type by Churn Status")
plt.show()

# 8. Payment Method Analysis
sns.countplot(x='Payment Method', data=customer_churn_df, hue='Customer Status')
plt.title("Payment Method by Churn Status")
plt.show()
```
